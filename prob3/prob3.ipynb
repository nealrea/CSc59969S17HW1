```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats as stats
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import matplotlib

#Numpy option
np.random.seed(78)

#Pandas Options
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", 400)

# For Matplotliab Visualization
matplotlib.rcParams['image.cmap'] = 'viridis'
%matplotlib inline
matplotlib.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (10, 6)

for x in range(0,4):
    if x == 0:
        df = pd.read_excel("/home/neal/Documents/USFoodCommodity.xlsx", sheetname = "94-98 FAH")
    if x == 1:
        temp = pd.read_excel("/home/neal/Documents/USFoodCommodity.xlsx", sheetname = "03-04 FAH")
        df = pd.concat([df,temp], axis = 1, ignore_index = True)
    if x == 2:
        temp = pd.read_excel("/home/neal/Documents/USFoodCommodity.xlsx", sheetname = "05-06 FAH")
        df = pd.concat([df,temp], axis = 1, ignore_index = True)
    if x == 3:
        temp = pd.read_excel("/home/neal/Documents/USFoodCommodity.xlsx", sheetname = "07-08 FAH")
        df = pd.concat([df,temp], axis = 1, ignore_index = True)

#Fruit types over time (Men)
applesAsFruitMen = np.array([df.iloc[78][7], df.iloc[78][23], df.iloc[78][39], df.iloc[78][55]])
bananasMen = np.array([df.iloc[80][7], df.iloc[80][23], df.iloc[80][39], df.iloc[80][55]])
berriesMen = np.array([df.iloc[81][7], df.iloc[81][23], df.iloc[81][39], df.iloc[81][55]])
grapesMen = np.array([df.iloc[82][7], df.iloc[82][23], df.iloc[82][39], df.iloc[82][55]])
melonsMen = np.array([df.iloc[83][7], df.iloc[83][23], df.iloc[83][39], df.iloc[83][55]])
orangesMen = np.array([df.iloc[84][7], df.iloc[84][23], df.iloc[84][39], df.iloc[84][55]])
citrusMen = np.array([df.iloc[87][7], df.iloc[87][23], df.iloc[87][39], df.iloc[87][55]])
stoneFruitMen = np.array([df.iloc[88][7], df.iloc[88][23], df.iloc[88][39], df.iloc[88][55]])
tropicalFruitMen = np.array([df.iloc[89][7], df.iloc[89][23], df.iloc[89][39], df.iloc[89][55]])

years = np.array([0,1,2,3])
ticks = ["1994-1998", "2003-2004", "2005-2006", "2007-2008"]

fig, ax = plt.subplots()
plt.figtext(0.5,0,"U.S. Food Commodity Availability by Food Source, 1994-2008",horizontalalignment = 'center',fontsize = 12)
plt.figtext(0.5,-0.05,"https://www.ers.usda.gov/publications/pub-details/?pubid=81817",horizontalalignment = 'center',fontsize = 12)
plt.xticks(years,ticks)
ax.plot(years, applesAsFruitMen, 'r', label = 'Apples as fruit')
ax.plot(years, bananasMen, 'y', label = 'Bananas')
ax.plot(years, berriesMen, 'b', label = 'Berries')
ax.plot(years, grapesMen, 'purple', label = 'Grapes')
ax.plot(years, melonsMen, 'green', label = 'Melons')
ax.plot(years, orangesMen, 'orange', label = 'Oranges, Total')
ax.plot(years, citrusMen, 'yellow', label = 'Other citrus fruit')
ax.plot(years, stoneFruitMen, 'fuchsia', label = 'Stone fruit')
ax.plot(years, tropicalFruitMen, 'dodgerblue', label = 'Tropical Fruit')
ax.set_ylim(ymin = 60)
ax.set_xlabel('time (year)')
ax.set_ylabel('mean consumption (percentage)')
ax.set_title('Fruit Types Over Time (Men)')
plt.legend(loc = 'lower left')
plt.show()
```

We can observe that while most fruits have seen an average rise in consumption, citrus fruit consumption (not including oranges or tropical fruits) has declined drastically from 1994-2008.  We can also note that stone fruit, apples, melons, and tropical fruit are the only fruits to increase in consumption from 2006-2008.
